```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
erf = pd.read_csv('https://github.com/QuantEcon/ShenzhenWinterCamp/raw/master/Jiao%20Shi/ER_predict.csv')
erf
```

## UIP

```python
erf['lner'] = np.log(erf[['CNYUSD']])
erf['dep'] = 100*erf['lner'].diff(periods = 1)
```

```python
erf['dep_1'] = 0
size_sample = 149
for i in range(size_sample):
    erf.loc[i,'dep_1'] = erf.loc[i+1,'dep']
erf.loc[size_sample,'dep_1'] = erf.loc[0,'dep']
```

```python
erf['int_diff'] = erf['CHN_IBR'] - erf['US_FFR']
erf['const'] = 1
```

```python
size_sample = 149
size_subsample = 60
fcst_int = np.zeros(size_sample - size_subsample)

for i in range (size_sample - size_subsample):
    reg_intsub = sm.OLS(endog=erf.loc[i:size_subsample-1+i,'dep_1'], exog=erf.loc[erf.index[i:size_subsample+i],['const', 'int_diff']], missing='drop')
    results_sub = reg_intsub.fit()
    fcst_int[i] = results_sub.predict(exog = [1, erf.loc[size_subsample+i,'int_diff']])
fcst_int
```

```python
dire1 = np.zeros(size_sample - size_subsample) 
real = np.zeros(size_sample - size_subsample)
for i in range(size_sample - size_subsample):
    real[i] = erf.loc[size_subsample+i, 'dep_1']
    dire = real[i]*fcst_int[i]
    if dire > 0:
        dire1[i]=1
percentage1 = dire1.sum()/(size_sample - size_subsample)
percentage1       
```

## The Taylor-rule model 

```python
erf['inflation_diff'] = erf['CHN CPI inflation'] - erf['USA CPI inflation']
predict_tr = np.zeros(size_sample - size_subsample)
erf['const'] = 1
for i in range (size_sample - size_subsample):
    y_chn = erf['CHN PMI'] - np.mean(erf.loc[i:size_subsample-1+i, 'CHN PMI'])
    y_us = erf['USA PMI'] - np.mean(erf.loc[i:size_subsample-1+i, 'USA PMI'])
    erf['y_diff'] = y_chn - y_us
    
    reg_trsub = sm.OLS(endog=erf.loc[i:size_subsample-1+i,'dep_1'], exog=erf.loc[erf.index[i:size_subsample+i],['const', 'inflation_diff','y_diff']], missing='drop')
    results_sub = reg_trsub.fit()    
    predict_tr[i] = results_sub.predict(exog = [erf.loc[size_subsample+i,['const', 'inflation_diff','y_diff']]])
predict_tr    
```

```python
dire2 = np.zeros(size_sample - size_subsample) 
real1 = np.zeros(size_sample - size_subsample)
for i in range(size_sample - size_subsample):
    real1[i] = erf.loc[size_subsample+i, 'dep_1']
    dire = real[i]*predict_tr[i]
    if dire > 0:
        dire2[i]=1
percentage2 = dire2.sum()/(size_sample - size_subsample)
percentage2       
```

## the random walk model

```python
dire3 = np.zeros(size_sample)
for i in range(size_sample):
    if erf.loc[i,'dep_1'] * erf.loc[i,'dep'] > 0:
        dire3[i]=1
percentage3 = dire3.sum()/size_sample
percentage3  
```
